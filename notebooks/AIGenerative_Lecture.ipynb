```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("book.pdf")

data = loader.load()
```

```python
data
```

```python
print (f'You have {len(data)} documents in your data')
print (f'''There are ~{np.mean([len(x.page_content) for x in data])} characters per document''')
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Evenly splitting the text documents to have Equal size documents
# Overlapping to ensure no connection is missed
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=400)

texts = text_splitter.split_documents(data)
```

```python
len(texts)
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings.openai import OpenAIEmbeddings

embedder = OpenAIEmbeddings(openai_api_key= api_key)
```

```python
# Storing these Vectors in memory database
vector_db = Chroma.from_documents(texts, embedder)
```

```python
#just texting the format
text = "This is a test document."
query_result = embedder.embed_query(text)
query_result[:5]
```

```python
query = "How do I establish a Class?"
docs = vector_db.similarity_search(query, k = 5)
```

```python
docs
```

## Let's summarize it

```python
# The QA Chain allows to ask questions over a document

from langchain.llms import OpenAI # Choosing which LLM
from langchain.chains.question_answering import load_qa_chain
```

```python
llm = OpenAI(temperature=0, openai_api_key=api_key) # finetuning the LLM
chain = load_qa_chain(llm, chain_type="map_reduce") #verbose=True to check under the hood
```

```python
query = "How does the author recommend I keep studying after the book?"
docs = vector_db.similarity_search(query)
```

```python
docs
```

```python
print(chain.run(input_documents=docs, question=query))

# Output: 'The author recommends continuing to study by doing additional exercises,
# case studies, and using Sawmpy, ...'
```
